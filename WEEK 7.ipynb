# POS TAGGING (PART OF SPEECH)

```python
nltk.download("averaged_perceptron_tagger")
```

```python
s = "i am going to universty"
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
tag_w = nltk.pos_tag(word_tokenize(s))
print(tag_w)
```

# CHUNKING

```python
# it take pos tag as input and return chunk as output
```

```python
grammer1 = "NP:{<DT>?<JJ>*<NN>}"
p = nltk.RegexpParser(grammer1)   
o = p.parse(tag_w)
print(o)
# o.draw()
```

# CHINKING

```python
grammer2 = r""" NP :{<.*>+}
                    }<JJJ>+{"""
p = nltk.RegexpParser(grammer2)
o = p.parse(tag_w)
print(o)
# o.draw()
```

# NER (NAMED ENTITY RECOGNITION)

```python
import nltk
from nltk import word_tokenize
nltk.download('maxent_ne_chunker')
```

```python

nltk.download('words')
```

```python
data = "Mr. Khan made a deal on a beach of Switzerland near WHO."
w = word_tokenize(data)
for i in w:
    tag_w=nltk.pos_tag(w)
N_E_R = nltk.ne_chunk(tag_w, binary=False)
print(N_E_R)
#N_E_R.draw()
```
