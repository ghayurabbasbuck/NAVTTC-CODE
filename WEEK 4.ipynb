# LINE PLOT

```python
from matplotlib import pyplot as plt
import numpy as np
plt.plot([2,4,6,8],[10,3,20,4])
plt.show()
```

# TWO LINE PLOT TOGETHER

```python
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x1,y1,"g",label = "line 1")
plt.plot(x2,y2,"b",label = "line 2")
plt.title("Two Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid(True)
plt.show()
```

# AREA PLOT

```python
from matplotlib import pyplot as plt
days = [1,2,3,4,5]
sleeping = [5,6,4,2,9]
eating = [3,6,7,5,7]
working = [7,6,8,5,4]
playing = [8,7,8,9,13]
plt.plot([],[],color="r",lw = 2, label="sleeping")
plt.plot([],[],color="g",lw = 2, label="eat")
plt.plot([],[],color="k",lw = 2, label="working")
plt.plot([],[],color="b",lw = 2, label="play")
plt.stackplot(days,sleeping,eating,working,playing,colors=["r","g","k","b"])
plt.legend()
plt.show()
```

# PIE PLOT

```python
slices = [7,2,2,13]
act = ["sleep","eat","work","play"]
cols = ["c","b","k","m"]
plt.pie(slices,labels=act,colors =cols,startangle = 90)
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
act = ["sleep","eat","work","play"]
cols = ["c","b","k","m"]
plt.pie(slices,labels=act,colors =cols,startangle = 90,shadow = True,explode = (0.1,0,0,0), autopct = '%1.1f%%')
plt.legend()
plt.show()
```

```python
# event1 = heat
# event 2 = face card
# event 3 = queen of heart

def event_probability(event,sample_space):
    probabilty=(event/sample_space)
    return(round(probabilty,2))
cards = 52
heart = 13
face_card=12
queen_heart = 1
heart_event = event_probability(heart,cards)
face_card_event = event_probability(face_card,cards)
queen_heart_event = event_probability(queen_heart,cards)
print(heart_event)
print(face_card_event)
print(queen_heart_event)
```

```python
# in percentage
# event1 = heat
# event 2 = face card
# event 3 = queen of heart

def event_probability(event,sample_space):
    probabilty=(event/sample_space)*100
    return(round(probabilty,2))
cards = 52
heart = 13
face_card=12
queen_heart = 1
heart_event = event_probability(heart,cards)
face_card_event = event_probability(face_card,cards)
queen_heart_event = event_probability(queen_heart,cards)
print(str(heart_event)+"%")
print(str(face_card_event)+"%")
print(str(queen_heart_event)+"%")
```
